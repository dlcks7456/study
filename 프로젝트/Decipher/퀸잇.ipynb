```python
import pandas as pd
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 방지

```

```python
df = pd.read_excel('220124.xlsx', index_col='record')
df
```

```python
# 광고 만족도

videonumber = 1
videovar = 'B1_%sr1'%(videonumber)

colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff']
explode = [0.05] * len(colors)

values = [ len(df[df[videovar]==i]) for i in range(1,6)]
labels = ['전혀 마음에 들지 않는다','마음에 들지 않는다', '보통이다', '마음에 드는 편이다', '매우 마음에 든다']

wedeprops = {'width':0.8, 'edgecolor':'w', 'linewidth':3}

def custom_autopct(pct) :
    return ('%.1f%%' % pct) if pct > 10 else ''

plt.figure(figsize=(5,7))
plt.pie(values, labels=labels, autopct=custom_autopct, counterclock=False, colors=colors, wedgeprops=wedeprops)
plt.title('광고%s : 얼마나 마음에 드시나요?'%(videonumber))
plt.show()
```

```python
def custom_autopct(pct) :
    return ('%.1f%%' % pct) if pct > 10 else ''

colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff']
explode = [0.05] * len(colors)
labels = ['전혀 마음에 들지 않는다','마음에 들지 않는다', '보통이다', '마음에 드는 편이다', '매우 마음에 든다']

#광고 1 ~ 8 전체 만족도 비교
fig, axs = plt.subplots(2, 4, figsize=(20, 10))
fig.suptitle('퀸잇 광고 만족도')

videonumber = 0
for i in range(0, 2) :
    for j in range(0, 4) :
        videonumber += 1
        videovar = 'B1_%sr1'%(videonumber)

        values = [ len(df[df[videovar]==i]) for i in range(1,6)]

        axs[i, j].pie(values, autopct=custom_autopct, counterclock=False, colors=colors, wedgeprops=wedeprops)
        axs[i, j].set_title('광고%s'%(videonumber))
        
fig.legend(labels, ncol=5, loc='lower center')
plt.savefig('전체 B1.png', dpi=100)
plt.show()
```

```python
# 연령 별 광고 1 ~ 8 전체 만족도 비교 (40대)
fig, axs = plt.subplots(2, 4, figsize=(20, 10))
fig.suptitle('40대(40~49) 퀸잇 광고 만족도')

videonumber = 0
for i in range(0, 2) :
    for j in range(0, 4) :
        videonumber += 1
        videovar = 'B1_%sr1'%(videonumber)

        values = [ len(df[ (df[videovar]==i) & (df['SQ4'].between(40,49)) ]) for i in range(1,6)]

        axs[i, j].pie(values, autopct=custom_autopct, counterclock=False, colors=colors, wedgeprops=wedeprops)
        axs[i, j].set_title('광고%s'%(videonumber))
        
fig.legend(labels, ncol=5, loc='lower center')
plt.savefig('40대 B1.png', dpi=100)
plt.show()
```
